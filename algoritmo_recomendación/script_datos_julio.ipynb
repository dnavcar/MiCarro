```python
import json

# --- Configuración ---
archivo_entrada = "mercadona_smartcart_dataset.jsonl"        # JSON con tickets originales
archivo_salida = "dataset_julio.json"    # JSON convertido con pares

tickets = []
pares = []

# --- 1. Leer tickets desde JSONL ---
with open(archivo_entrada, "r", encoding="utf-8") as f:
    for line in f:
        t = json.loads(line)
        productos_unicos = sorted(list(set(t["products"])))  # eliminar duplicados y ordenar
        tickets.append(productos_unicos)

        # Generar pares (entrada → siguiente producto)
        for i in range(1, len(productos_unicos)):
            entrada = sorted(productos_unicos[:i])  # orden determinista
            siguiente = productos_unicos[i]
            pares.append({"entrada": entrada, "siguiente": siguiente})

# --- 2. Guardar en JSON ---
with open(archivo_salida, "w", encoding="utf-8") as f:
    json.dump({
        "tickets": tickets,
        "pares": pares
    }, f, ensure_ascii=False, indent=2)

print(f"Archivo generado: {archivo_salida}")
print(f"Total tickets: {len(tickets)}, Total pares: {len(pares)}")
```
